```python
# EXPLORATION
ax = df['Activity'].value_counts().plot(kind='bar', title='Data by activity type');
ax.set_xlabel("Activity")
ax.set_ylabel("Number Of Records")
```

```python
ax = df['UserName'].value_counts().plot(kind='bar', title='Data examples by user');
ax.set_xlabel("User Names")
ax.set_ylabel("Number Of Records")
```

```python
def plot_activity(activity, df):
    data = df[df['Activity'] == activity][['AccX', 'AccY', 'AccZ']][:200]
    axis = data.plot(subplots=True, figsize=(16, 12), 
                     title=activity)
#     axis.set_xlabel("Count")
#     axis.set_ylabel("Accelerometer Reading")
    for ax in axis:
        ax.legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
```

```python
plot_activity("Sitting", df)
```

```python
plot_activity("Walking", df)
```

```python
plot_activity("Upstairs", df)
```

```python
plot_activity("Downstairs", df)
```
